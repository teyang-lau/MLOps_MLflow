```python
schema = {}
for col in data.columns:
    dt = data[col].dtype
    if dt.kind in 'iufc':
        schema[col] = {"type": str(dt), "min": float(data[col].min()), "max": float(data[col].max())}
    else:
        schema[col] = {"type": str(dt), "domain": list(set(data[col]))}

schema
```

```python
with open("schema.json", "r") as read_file:
   schema1 = json.load(read_file)
schema1
```

```python
import warnings
data1 = data.copy()
data1.at[0, "resale_price"] = 5000000
# data1['resale_price'] = data1['resale_price'].astype(str)
data1.at[0, "flat_type"] = "hoho"
data1.at[1, "flat_type"] = "hohe"
for col in data1.columns:
    if col not in schema1:
        print("Column {} is missing from dataset".format(col))
        break
    dt = data1[col].dtype
    if dt.kind in 'iufc':
        if dt != schema1[col]["type"]:
            print("Expected column `{}` to be \"{}\" but got \"{}\"".format(col, schema1[col]["type"], dt))
            print("Terminating data validation ...")
            break
        d_max, d_min = data1[col].max(), data1[col].min()
        if d_max > schema1[col]["max"] :
            warnings.warn("Column `{}` has values ({}) higher than max of schema ({})".format(col, d_max, schema1[col]["max"]))
        if d_min < schema1[col]["min"]:
            warnings.warn("Column `{}` has values ({}) lower than min of schema ({})".format(col, d_min, schema1[col]["min"]))
            break
    else: # if object
        if dt != schema1[col]["type"]:
            print("Expected column `{}` to be \"{}\" but got \"{}\"".format(col, schema1[col]["type"], dt))
            print("Terminating data validation ...")
            break
        old_set = set(schema1[col]["domain"])
        diff = set(data1[col]).difference(old_set)
        if diff:
            print("Column `{}` domain contains \"{}\" that are not present in schema".format(col, diff))
            print("Terminating data validation ...")
            break
```
